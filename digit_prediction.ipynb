```python
# multiclass classification
#Logistic Linear regression
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
dir(digits)
```

```python
#numeric data
digits.data[0]#image represented by 8x8 dimensions
```

```python
#image
plt.gray()
for i in range (5):
    plt.matshow(digits.images[i])
```

```python
digits.target[0:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python

X_train, X_test, y_train, y_test = train_test_split(digits.data,digits.target,test_size=0.20)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
plt.matshow(digits.images[67])
```

```python
digits.target[67]
```

```python
model.predict([digits.data[67]])
```

```python
model.predict(digits.data[0:5])
```

```python
y_predict = model.predict(X_test)
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(y_test,y_predict)
cm
```

```python
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
# 41 indicates 41 times when data was 1 theprediction was 1
# 1 in column 1 row 4 means that 1 time data was4 and it predicted 1
```
